```python
df["price"] = df["price"].astype(float)
```

```python
df.dropna()
```

#features and prices Divide



```python
x = df.drop("price",axis = 1)
y = df["price"]
```

## Data preprocessing

```python
x = pd.get_dummies(x,drop_first=True)
```

```python
x
```

```python
x.info()
```

# Train and test data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

##Data Standardizer

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

# Model Selection

## Linear Regression

###train Concat

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

train_combined = pd.concat([x_train, y_train], axis=1)

train_combined_cleaned = train_combined.dropna()

x_train_cleaned = train_combined_cleaned.drop(columns=['price'])
y_train_cleaned = train_combined_cleaned['price']

model.fit(x_train_cleaned, y_train_cleaned)
```

```python
x_train_cleaned
```

### Test Concat



```python
test_combined = pd.concat([x_test,y_test], axis = 1)
test_combined_cleaned = test_combined.dropna()
x_test_cleaned = test_combined_cleaned.drop(columns=['price'])
y_test_cleaned = test_combined_cleaned['price']
```

Line

###Prediction for the x_text

```python
y_pred = model.predict(x_test_cleaned)
```

### MAE, MSE, r2_score

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("MAE:", mean_absolute_error(y_test_cleaned, y_pred))
print("MSE:", mean_squared_error(y_test_cleaned, y_pred))
print("R2 Score:", r2_score(y_test_cleaned, y_pred))
```

###**Coefficients**

```python
coefficients = pd.DataFrame({
    "Feature": x_train_cleaned.columns,
    "Coefficient": model.coef_
})

coefficients
```

```python
print("Intercept:", model.intercept_)

for feature, coef in zip(x_train_cleaned.columns, model.coef_):
    print(f"{feature}: {coef}")
```

#price Comparision between models

```python
import matplotlib.pyplot as plt

models = ["Linear Regression", "Decision Tree", "Random Forest"]
prices = [predicted_price_lr, predicted_price_dt, predicted_price_rf]

plt.figure(figsize=(8,5))
for model, price in zip(models, prices):
  plt.bar(model, price)

plt.title("Predicted Car Price Comparison")
plt.xlabel("Models")
plt.ylabel("Predicted Price ($)")

plt.show()
```

#MAE Comparision

```python
models = ["Linear Regression", "Decision Tree", "Random Forest"]
mae_values = [1272, 1532, 1136]

plt.figure(figsize=(8,5))
for model, mae in zip(models, mae_values):
  plt.bar(model, mae)

plt.title("MAE Comparison Between Models")
plt.xlabel("Models")
plt.ylabel("Mean Absolute Error")

plt.show()
```

# R2 Score Comparision

```python
models = ["Linear Regression", "Decision Tree", "Random Forest"]
r2_scores = [0.91, 0.87, 0.924]

plt.figure(figsize=(5,6))
for model, r2_score in zip(models, r2_scores):
  plt.bar(model, r2_score)

plt.title("R² Score Comparison")
plt.xlabel("Models")
plt.ylabel("R² Score")

plt.show()
```
